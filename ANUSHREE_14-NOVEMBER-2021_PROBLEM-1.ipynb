Outlier treatment:
There are nearly no outliers in most of the numerical columns, only outlier is in economic.cond.national and economic.cond.household variable. In Gaussian Naive Bayes, outliers will affect the shape of the Gaussian distribution and have the usual effects on the mean etc. So depending on our use case, it makes sense to remove outlier .

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
lr,ur=remove_outlier(df_new["economic.cond.national"])
df_new["economic.cond.national"]=np.where(df_new["economic.cond.national"]>ur,ur,df_new["economic.cond.national"])
df_new["economic.cond.national"]=np.where(df_new["economic.cond.national"]<lr,lr,df_new["economic.cond.national"])
```

```python
lr,ur=remove_outlier(df_new["economic.cond.household"])
df_new["economic.cond.household"]=np.where(df_new["economic.cond.household"]>ur,ur,df_new["economic.cond.household"])
df_new["economic.cond.household"]=np.where(df_new["economic.cond.household"]<lr,lr,df_new["economic.cond.household"])
```
